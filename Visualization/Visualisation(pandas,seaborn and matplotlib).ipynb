```python
warnings.filterwarnings('ignore')
```

Scatter plot

```python
# To find the distribution between two numerical columns
# 

x=tips['total_bill']
y=tips['tip']
```

```python
plt.scatter(x=x,y=y,c='green')
```

```python
# s- shape
plt.title('Scatter Plot')
plt.xlabel('tip')
plt.ylabel('total_bill')
plt.scatter(x=x,y=y,s=y**2, c='green')
```

Bar Plot

```python
# Mainly for Categorical variable
# to see the count
# rot =0 horizontal label

tips['smoker'].value_counts().plot(kind='bar')
```

```python
tips['smoker'].value_counts().plot(kind='bar',rot=0)
```

```python
tips['smoker'].value_counts().plot(kind='bar',rot=0)
plt.text(x=0,y=10,s=tips['smoker'].value_counts()[0])
plt.text(x=1, y=11,s=tips['smoker'].value_counts()[1])
plt.title('Bar Plot')
plt.xlabel('Smoker')
plt.ylabel('Count')
```

```python
tips.day.value_counts()
```

```python
# This is for Categorical variables
# autopct is show in % format with one decimal value
plt.pie(tips.day.value_counts(),labels=['Sat','Sun','Thur','Fri'],autopct='%.1f%%')
```

```python
plt.pie(tips.day.value_counts(),labels=['Sat','Sun','Thur','Fri'],autopct='%.1f%%',explode=[0,0,0,0.2],radius=2)
```

Histogram plot

```python
# it is only for the numerical variables.
# it is showing the data in the specfic range/bucket
# number of range/bars can be defined as bins
# its mainly to see the distribution of the data

plt.hist(tips['tip'],bins=5)
```

```python
tips['tip'].plot(kind='hist')
tips['tip'].plot(kind='density')
plt.xlabel('tip')
```

Box Plot

```python
# Box plot is for to see the 5 summary values
# Mainly contributing to outlier detection
plt.boxplot(tips['total_bill'])
plt.title('Box Plot')
```

```python
tips['total_bill'].plot(kind='box')
```

```python
#Conditional Box plot
tips[['total_bill','tip']].plot(kind='box')
```

```python
tips[['tip']].plot(kind='box')
```

Stirp Plot

```python
# Categorical vs numerical 
# Similar to bar plot
sns.stripplot(x='day',y='tip',data=tips)
```

```python
sns.barplot(x='day',y='tip',data=tips,hue='sex')
```

Count Plot

```python
sns.countplot(x='day',data=tips,hue='sex')
```

Swarm Plot

```python
# Swarm plot also similar to strip plot
# design look will be chnaging slightly
sns.swarmplot(x='day',y='tip',data=tips)
```

```python
sns.swarmplot(x='day',y='tip',data=tips,hue='sex')
```

Violin Plot

```python
#Similar to Swarmplot
sns.violinplot(x='day',y='tip',data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='tip',data=tips,hue='sex',split=True)
```

Pair Plot

```python
#Brings relationship between all the numerical variables
sns.pairplot(data=tips)
```

Distribution Plot

```python
# Combination of histogram and density
# We can understand the distribution of the population
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],hist=False)
```

```python
for time in ['Lunch','Dinner']:
    df=tips[tips['time']==time]
    sns.distplot(df['total_bill'],hist=False,kde=True
                ,label=time)
```

Correlation - Heatmap

```python
sns.heatmap(tips.corr(),annot=True)
```
